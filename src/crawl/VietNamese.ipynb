```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
'''with open('../../resource/Japan_artist.json') as user_file:
    file_contents = user_file.read()
    parsed_json = set(json.loads(file_contents))
'''
newList = set()
```

```python
def get_all_similar_artist(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        page_content = response.text
        soup = BeautifulSoup(page_content, 'html.parser')
        #print(soup)
        similar_artists = soup.select(".big-artist-list-title > a")
        #print(similar_artists)
        for artist in similar_artists:
            if "," in artist or "&" in artist:
                continue
            newList.add(artist.get_text().strip())
    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")
        
for i in range(1, 66):
    print(i, end=' ')
    get_all_similar_artist(f"https://www.last.fm/tag/vietnamese/artists?page={i}")
```

```python
result = filter(lambda artist: ',' not in artist, newList)
result = filter(lambda artist: '&' not in artist, result)
result = filter(lambda artist: '-' not in artist, result)
result = filter(lambda artist: 'ft' not in artist, result)
result = list(result)
```

```python
len(result)
```

```python
result.sort()
with open("artist_VietNam.json", "w+") as outfile:
     json.dump(result, outfile)
```
